## Описание проекта

Даны данные сервиса Яндекс.Афиша с июня 2017 по конец мая 2018 года:
* лог сервера с данными о посещениях сайта Яндекс.Афиши,
* выгрузка всех заказов за этот период,
* статистика рекламных расходов.

Для того, чтобы снизить расходы, отказаться от невыгодных источников трафика и перераспределить бюджет, необходимо изучить:
* как клиенты пользуются сервисом,
* когда делают первые покупки на сайте,
* сколько денег приносит компании каждый клиент,
* когда расходы на привлечение клиента окупаются.

### Когортный анализ

Подготовим данные. Сгруппируем данные по столбцу uid и для каждого клиента получим дату его первого заказа.
Добавим эту информацию в таблицу заказов и выделим из столбцов first_order_date и order_date месяцы.
Аналогично для таблицы посещений visits

```python
def make_first_date_column(data, first_name, col_name, first_month_name, month_name):
    first_date_by_user = data.groupby('uid')[col_name].min() 
    first_date_by_user.name = first_name
    new_data = data.merge( first_date_by_user, on='uid' )
    
    new_data[first_month_name] = new_data[first_name].astype('datetime64[M]')
    new_data[month_name] = new_data[col_name].astype('datetime64[M]')
    return new_data
    

orders = make_first_date_column(orders, 'first_order_date', 'order_date', 'first_order_month', 'order_month' )
display(orders.head(5))
```

```python
visits = make_first_date_column(visits, 'first_visit_date', 'start_ts', 'first_visit_month', 'visit_month' )
display(visits.head(5))
```

```python
cohort_grouped = orders.groupby('first_order_month').agg({'uid':'nunique','revenue':'sum'})
display(cohort_grouped) 
```

#### DAU, WAU, MAU

```python
def my_hist(data,param):
    data.plot(title=param['title'], figsize=(12,3), kind='hist', grid='True', bins = param['bins'])
    try:
        plt.axvline(x=int(data.median()), linewidth=2, color='m',linestyle='--')
    except:     
        plt.xlabel('Значение')
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.show()
    
def my_bar(data,param):
    ax = data.plot(title=param['title'], figsize=(12,3), kind='bar', grid='True')
    plt.xlabel(None)    
    plt.ylabel(param['ytitle'])
    plt.show()
    
def my_plot(data, param):
    data.plot(figsize=(12, 4), title=param['title'])
    plt.ylabel(param['xtitle'])
    plt.xlabel(param['ytitle'])
    plt.legend(title=param['ltitle'], loc="center right", bbox_to_anchor=(0.7, 0, 0.5, 1))
    if 'xtick' in param:
        plt.xticks([])
        plt.show()
    else:
        plt.show()
        
def my_heatmap(data, param):
    sns.set(style='white')
    plt.figure(figsize=(13, 9))
    if 'yticklabels' in param:
        sns.heatmap(data, vmin=param['vmin'], vmax=param['vmax'], annot=True, \
                    yticklabels = param['yticklabels'],\
                    fmt=param['fmt'], linewidths=1, linecolor='gray')
    else:
        sns.heatmap(data, vmin=param['vmin'], vmax=param['vmax'],annot=True,\
                    fmt=param['fmt'], linewidths=1, linecolor='gray')
    plt.title(param['title'])
    plt.ylabel(param['ytitle'])
    plt.xlabel(param['xtitle'])
    plt.show()
    
visits['session_year']  = visits['start_ts'].dt.year
visits['session_month'] = visits['start_ts'].dt.month
visits['session_week']  = visits['start_ts'].dt.week
visits['session_date']  = visits['start_ts'].dt.date

dau_total = visits.groupby('session_date').agg({'uid': 'nunique'})
wau_total = visits.groupby(['session_year', 'session_week']).agg({'uid': 'nunique'})
mau_total = visits.groupby(['session_year', 'session_month']).agg({'uid': 'nunique'})

my_hist(dau_total,{'title':'Количество новых пользователей в день', 'bins':50})
my_hist(wau_total,{'title':'Количество новых пользователей в неделю', 'bins':50})
my_hist(mau_total,{'title':'Количество новых пользователей в месяц', 'bins':50})

print('DAU:',int(dau_total.mean()))
print('WAU:',int(wau_total.mean()))
print('MAU:',int(mau_total.mean()))
```

#### Retention Rate. Cредний Retention Rate на второй месяц «жизни» когорт

В таблицу visits уже добавлены необходимые нам столбцы с месяцем заказа ('visit_month') и месяцем первого заказа ('first_visit_month') для пользователя. Найдем разницу между ними в целом числе месяцев - время жизни когорты.
Необходимо округлять перед приведением к целому типу, так как в np.timedelta64(1,'M') средний объем месяца, и мы теряем, например, 1 месяц 2017-06-01 когорты целиком

```python
visits['cohort_lifetime'] = visits['visit_month'] - visits['first_visit_month']
visits['cohort_lifetime'] = visits['cohort_lifetime'] / np.timedelta64(1,'M')
visits['cohort_lifetime'] = round(visits['cohort_lifetime']).astype('int') 
display(visits.sample(5))
```

Посчитаем для каждой когорты количество активных пользователей на определённую «месяц жизни»

```python
cohorts = visits.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
display(cohorts.head(5))
```

Найдём исходное количество пользователей в когорте. Поделим текущее число пользователей на изначальное число пользователей и найдем Retention rate

```python
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'}) 

cohorts = cohorts.merge(initial_users_count,on='first_visit_month')
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']

retention_pivot = cohorts.pivot_table(index='first_visit_month',
                                      columns='cohort_lifetime',
                                      values='retention',
                                      aggfunc='sum')
display(retention_pivot)
```

```python
retention_pivot_norm = pd.DataFrame(retention_pivot.to_records())
retention_pivot_norm['first_visit_month'] = retention_pivot_norm['first_visit_month'].dt.strftime('%Y-%m') 
my_heatmap(retention_pivot, {'vmax' : 0.1,
                             'vmin' : 0,
                                    'fmt' : '.1%',
                                    'title' : 'Retention rate',
                                    'xtitle' : 'Lifetime',
                                    'ytitle' : 'Corort',
                                    'yticklabels' : retention_pivot_norm['first_visit_month'].to_list()
                                    })
```

#### Изменение во времени LTV покупателя по когортам. Средний  LTV по когортам за 6 месяцев

Количество уникальных покупателей в каждой когорте. Создадим функцию, чтобы считать это число в зависимости от данных - вся таблица или срез по источнику

```python
def cohort_sizes_by_data(data):
    first_orders = data.loc[:,['uid','first_order_month']]
    cohort_sizes = first_orders.groupby('first_order_month').agg({'uid': 'nunique'}).reset_index()
    cohort_sizes.columns = ['first_order_month', 'n_buyers']
    return cohort_sizes

display( cohort_sizes_by_data(orders) )
```

* Сгруппируем заказы в когорты. 
* Посчитаем валовую прибыль
* Найдем возраст каждой когорты в месяцах (целое число)
* LTV когорт
* Средний LTV

```python
margin_rate = 1

def report_df_by_data(data):
    cohorts = data.groupby(['first_order_month','order_month']).agg({'revenue': 'sum'}).reset_index()
    cohort_sizes = cohort_sizes_by_data(data)
    report  = pd.merge(cohorts, cohort_sizes, on='first_order_month')
    report['age'] = (report['order_month'] - report['first_order_month']) / np.timedelta64(1, 'M') 
    report['age'] = report['age'].round().astype('int')
    report['ltv'] = (report['revenue'] * margin_rate) / report['n_buyers'] 
    return report

report = report_df_by_data( orders )    
ltv_pivot = report.pivot_table(
        index='first_order_month', 
        columns='age', 
        values='ltv', 
        aggfunc='mean').round(2)

ltv_pivot = ltv_pivot.cumsum(axis=1)
 
display(ltv_pivot)
```

```python
ltv_pivot_norm = pd.DataFrame(ltv_pivot.to_records())
ltv_pivot_norm['first_order_month'] = ltv_pivot_norm['first_order_month'].dt.strftime('%Y-%m')
my_heatmap(ltv_pivot, {'vmax' : 12,
                       'vmin' : 4,
                       'fmt' : '.2f',
                       'title' : 'LTV',
                       'xtitle' : 'Lifetime когорты',
                       'ytitle' : 'Когорта',
                       'yticklabels' : ltv_pivot_norm['first_order_month'].to_list()
                     })
```

```python
a = ltv_pivot_norm['first_order_month'].to_list()[:7]
print('Средний LTV за 6 месяцев после первой покупки: {:.2f}'.format(ltv_pivot.mean(axis=0)[5]))
display(ltv_pivot_norm.loc[ ltv_pivot_norm['first_order_month'].isin(a),['first_order_month','5']])
#display(ltv_pivot.mean(axis=0))
```

* В среднем каждый покупатель из сентябрьской когорты принес по 12.32 единиц валовой прибыли за 6 месяцев «жизни» когорты. Число получилось за счет максимального среднего чека в данной когорте за декабрь.
* Чтобы вложения в маркетинг окупались за 6 месяцев, привлечение одного покупателя не должно превышать 7.98 у.е.

#### Cредний CAC на одного покупателя для всего проекта и для каждого источника трафика.

Выделим основной источник для пользователя, например, тот, с которого был первый переход.

```python
first_source = visits.loc[visits['start_ts'] == visits['first_visit_date']][['uid','source_id']]
orders = orders.merge(first_source, on='uid', how='left')
```

* Выделим месяц из даты в таблице рекламных расходов. Сгруппируем расходы по месяцам.
* Объединим сгруппированную таблицу расходов с таблицей LTV из 4.3.4 
* Вычислим САС по когортам
* Вычислим САС по источникам трафика

```python
costs['month'] = costs['dt'].astype('datetime64[M]')
```

```python
def color_cac(value):
    try:
        if value > 7.98:
            color = 'red'
        else:
            color = 'black'
    except:
        color = 'black'
    return 'color: %s' % color

def color_romi(value):
    try:
        if value > 1:
            color = 'red'
        else:
            color = 'black'
    except:
        color = 'black'
    return 'color: %s' % color

def display_cac_romi(df, func):
    df = pd.DataFrame(df.to_records())
    if 'first_order_month' in df:
        df['first_order_month'] = df['first_order_month'].dt.date
    df = round(df,2)
    df = df.fillna('')
    display( df.style.applymap(func, subset=df.columns[1:]) )

def cac_report_by_costs(data_cost, data_orders):
    monthly_costs      = data_cost.groupby('month').agg({'costs': 'sum'}).reset_index() 
    report_data_orders = report_df_by_data(data_orders)
    report_cac         = pd.merge(report_data_orders, monthly_costs, left_on='first_order_month', right_on='month')
    report_cac['cac']  = report_cac['costs'] / report_cac['n_buyers']
    return report_cac

def cac_mean_report_by_costs(data_cost, data_orders):
    report_cac = cac_report_by_costs(data_cost, data_orders)
    report_cac_mean = report_cac.groupby('first_order_month')['cac'].mean()
    return report_cac_mean

print('Средний CAC на одного покупателя для всего проекта и для каждого источника')
report_all = cac_mean_report_by_costs(costs, orders)
report_all = report_all.to_frame()
report_all.columns = ['all']

for source_id in costs['source_id'].unique():
    costs_by_source = costs.query('source_id == ' + source_id.astype(str))
    orders_by_source = orders.query('source_id == ' + source_id.astype(str))
    report_by_source = cac_mean_report_by_costs(costs_by_source, orders_by_source )
    
    report_by_source = report_by_source.to_frame()
    report_by_source.columns = ['source ' + source_id.astype(str)]
    report_all = report_all.merge(report_by_source, on='first_order_month')
    
display_cac_romi(report_all, color_cac)
```

* 3 и 4 источники обходятся дороже, чем остальные
* На маркетинг больше всего было потрачено в августе, но это казалось бы оправдало себя в осенние месяцы - с октября по декабрь на сайте  большее количество покупателей, чем в другие месяцы. Затраты на привлечение одного покупателя в большинстве случаев оказались больше 7.98, те не окупились в последующие полгода

#### ROMI по когортам в разрезе источников. Окупаемость за одинаковые периоды жизни когорт. 

```python
def romi_report_by_costs(data_cost, data_orders):
    report_romi = cac_report_by_costs(data_cost, data_orders)
    report_romi['romi'] = report_romi['ltv'] / report_romi['cac']
    romi  = report_romi.pivot_table(
                    index='first_order_month', 
                    columns='age', 
                    values='romi', 
                    aggfunc='mean')
    romi = romi.cumsum(axis=1)
    display_cac_romi(romi, color_romi) 
    my_plot(romi, {'title':'ROMI в разные периоды "жизни" когорт',
                  'xtitle':'ROMI',
                  'ytitle':'Когорта',
                  'ltitle':'Возраст когорты'})
    
    my_heatmap(romi, {'vmax' : 1.1,
                      'vmin' : 0,
                      'fmt' : '.2f',
                       'title' : 'Romi',
                       'xtitle' : 'Возраст когорты',
                       'ytitle' : 'Когорта',
                        'yticklabels' : ltv_pivot_norm['first_order_month'].to_list()
                     })
    print('ROMI для источника', source_id,' для средней когорты')
    display_cac_romi(romi.mean(axis=0).to_frame(), color_romi) 
    return romi


for source_id in costs['source_id'].unique():
    print('ROMI по когортам для источника', source_id)
    costs_by_source = costs.query('source_id == ' + source_id.astype(str))
    orders_by_source = orders.query('source_id == ' + source_id.astype(str))
    report_by_source = romi_report_by_costs(costs_by_source, orders_by_source)
```
