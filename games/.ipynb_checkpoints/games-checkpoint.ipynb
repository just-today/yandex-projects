### Общее описание данных <a id="data_discribe"></a>


* Name — название игры
* Platform — платформа
* Year_of_Release — год выпуска
* Genre — жанр игры
* NA_sales — продажи в Северной Америке (миллионы проданных копий)
* EU_sales — продажи в Европе (миллионы проданных копий)
* JP_sales — продажи в Японии (миллионы проданных копий)
* Other_sales — продажи в других странах (миллионы проданных копий)
* Critic_Score — оценка критиков (максимум 100)
* User_Score — оценка пользователей (максимум 10)
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

Рейтинг:
* «EC» («Early childhood») — «Для детей младшего возраста»
* «E» («Everyone») — «Для всех» Первоначально "K-A" ("Kids to Adults")
* «E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»
* «T» («Teen») — «Подросткам»
* «M» («Mature») — «Для взрослых»
* «AO» («Adults Only 18+») — «Только для взрослых»

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import numpy as np
import math as math

try:
    df = pd.read_csv('https://code.s3.yandex.net/datasets/games.csv')
except:
    df = pd.read_csv('games.csv')

pd.set_option('display.max_columns',50) 
pd.set_option('display.max_rows',50)

def display_data(data,rows_num,description):
    if description:
        print(description)
    rows_num = rows_num or 5
    display(data.head(rows_num))
    print('Общая информация')
    data.info()
    null_values = data.isnull().sum()
    null_len =  len(null_values[null_values > 0])
    if null_len == 0:
        print('Сумма пропущенных значений: 0')
    else:
        print('Сумма пропущенных значений:')
        display(null_values[null_values > 0])
    print('Дубликатов строк: {:.0f}'.format(data.duplicated().sum()))
    print(' ')

display_data(df,5,'Таблица df с данными о продажах игр')
```

#### Вывод<a id="resume1"></a>

Каждая стока таблице содержит информацию об играх определенного жанра на определенной платформе с оценками критиков и пользователей, а также рейтингом от организации ESRB. 3 проблемы, которые нужно решить:
* В столбцах Name, Year_of_Release, Genre, Critic_Score, User_Score, Rating есть пропущенные значения
* Некачественные названия столбцов
* Год выхода суть целое число (заменим на int(8))
* Оценка пользователей - число с плавающей точкой

### Предобработка данных <a id="data_preprocessing"></a>

<b>Переименование столбцов</b> <a id="rename_cols"></a>

```python
df.set_axis(df.columns.str.lower(),axis='columns',inplace=True)
display(df.columns)
```

<b>Замена пропущенных значений</b> <a id="nan_values"></a>

Удалим игры, для которых отсутствуют название или жанр. Это очень старые игры, которые точно не будут учавствовать в рекламной компании.

```python
data = df.loc[df['name'].isnull() | df['genre'].isnull()]
display(data)
df.drop(data.index,inplace=True)
```

Для 1% данных отсутствует год выхода. Удалим эти данные без последствий для результата

```python
data = df.loc[df['year_of_release'].isnull()]
display(data.head())
df.drop(data.index,inplace=True)
```

Добавим столбец с суммой продаж 'sum_sales'. Для двух строк эта сумма оказалась равна 0, удалим.

```python
df['sum_sales'] = df['na_sales'] + df['jp_sales'] + df['eu_sales'] + df['other_sales'] 
display(df.loc[df['sum_sales'] == 0]) 
df.drop(df.loc[df['sum_sales'] == 0].index,inplace=True)
```

Гистограмма распределения

```python
def display_sum_sales(data):
    sum_sales_describe = data['sum_sales'].describe()
    display(sum_sales_describe)
    quant = data['sum_sales'].quantile([0.05, 0.95])
    data['sum_sales'].plot(title='Сумма проданных копий', figsize=(12,3), kind='hist', bins=70, grid='True')
    plt.axvline(x=sum_sales_describe['50%'], linewidth=2,color='m',linestyle='--')
    plt.axvline(x=sum_sales_describe['max'], linewidth=2, color='r',linestyle='--')
    plt.axvline(x=quant[0.95], linewidth=2, color='m',linestyle='--')
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.show()
    return sum_sales_describe

sum_sales_describe = display_sum_sales(df)
sum_sales_describe_10 = display_sum_sales(df.query('sum_sales < 10'))
```

Заменим tbd в 'user_score' на nan, тк то, что оценка будет добавлена, но не добавлена на момент выгрузки данных ничего не меняет для текущего исследования

```python
df.loc[df['user_score'] == 'tbd','user_score'] = np.NaN
```

Пользовательские оценки очень субъективны, и несмотря на корреляцию 0,5 не так уж сильно зависят от мнения критиков. Заменим на маркер.

```python
df['user_score'] = df['user_score'].fillna('-1')
df['critic_score'] = df['critic_score'].fillna(-1)
```

Рейтинг
Заменим K-A на Е, так как одно (K-A - устаревшее)
Замену Nan я не вижу, так как во всех жанрах могут быть игры разного рейтинга. Возможно, заменить на "неизвестен"

```python
df.loc[df['rating'] == 'K-A','rating'] = 'E'
display(df.loc[df['rating'].isnull()].sort_values(by='year_of_release', ascending=False).head(5))
df['rating']= df['rating'].fillna('unknown')
```

<b>Преобразование типов</b><a id='sec23'></a>

```python
df['year_of_release'] = df['year_of_release'].astype('int')
df['user_score'] = df['user_score'].astype('float')
df.info()
```

<b>Вывод</b><a id='result2'></a>

* Несколько строк с пропущенными значениями в столбцах с названием, годом выпуска и жанром игры. Пропущенных значений в этих столбцах менее 1%, поэтому их удаление никак не скажется на результате исследований. Возможно, это пропуски при загрузке данных из разных источников
* Пропущенные значения в оценках пользователей и критиков вполне естественны, так как не всякая компания, возможно, привлекает критиков для оценки каждой игры. Возможно, квалифицированных критиков в этой области не достаточно, чтобы охватить такое количество игр, ведь игру надо пройти, чтобы составить мнение. Оценки пользователей также дело добровольное, не каждый любит писать отзывы. Так как корреляция продаж и отзывов не столь большая, как видно будет из дальнейших исследований, то производители игр не так заинтересованы в стимуляции написания отзыва и предоставления интерфейса для этого пользователю.
* Нет корректной замены для пропущенных значений в рейтинге, так как во всех жанрах есть игры с райным рейтингом. От суммы продаж рейтинг тоже не линейно зависит, так как есть Покемоны, например, с большими продажами
* Так как нет возможности получить пропущенные значения для user_score, critic_score и rating из наших данных, а таких строк очень много, то заменили их на маркеры.

### Исследовательский анализ данных <a id="data_analiz"></a>

<b>Выпуски игр в разные годы</b>  <a id="data_analiz1"></a> <a id="gr"></a>

```python
def my_hist(data,param):
    data.plot(title=param['title'], figsize=(12,3), kind='hist', grid='True', bins = param['bins'])
    try:
        plt.axvline(x=data.median(), linewidth=2, color='m',linestyle='--')
    except:     
        plt.xlabel('Значение')
    plt.xlabel('Значение')
    plt.ylabel('Частота')
    plt.show()
    

def my_bar(data,title,ytitle):
    ax = data.plot(title=title, figsize=(12,3), kind='bar', grid='True')
    plt.xlabel(None)    
    plt.ylabel(ytitle)
    plt.show()
```

```python
data = df.pivot_table(index='year_of_release', values='name', aggfunc='count')
data.columns = ['Количество игр']
data.index.name = None
display(data,data.describe())
my_hist(data,{'title':'Количество новых игр в разные годы','bins':len(data)})
my_bar(data,'Количество новых игр в разные годы','Выпущено игр')
```

* До 1994 года выпускалось до 100 игр в год. Настоящее развитие интернета начинается в середине 90-х. Также персональные компьютеры доступны не всем и еще не стали предметом домашнего обихода. 
* C 1995 года количество выпускаемых игр непрерывно растет и достигает максимума в 2009 (1426), потом снова наблюдается спад интереса. Возможно, это связано с развитием  и большой популярностью соцсетей. В 2006 появляются Facebook и VK, но настоящего развития и многомиллионной аудитории они достигают как раз примерно к 2009 году. Возможно из-за роста вариантов интернет активностей для пользователя несколько снижается интерес именно к играм
* Возможности компьютерного "железа" тормозили развитие графики. С отрисовкой сложной графики стоимость игр сильно возросла, поэтому у производителей нет возможности выпускать так много игр, а у аудитории покупать

<b>Продажи игр по платформам</b>  <a id="data_analiz2"></a>

```python
def color_positive_red(val):
    color = 'white' 
    if val > 0:
        color = '#FFCC99'
    return 'background: %s' % color

def pivot_table_pl_sum_by_years(data):
    data1 = data.pivot_table(index='platform', columns ='year_of_release', values='sum_sales', aggfunc='sum')
    return data1.style.applymap(color_positive_red)

data = df.pivot_table(index='platform', columns ='year_of_release', values='sum_sales', aggfunc='sum')
display(pivot_table_pl_sum_by_years(df), data.describe())
```

- Посчитаем, сколько "живут" платформы
- Таблицу из мультииндексов сгруппируем снова по платформам
- Посмотрим на гистограмму и среднюю продолжительность "жизни"

```python
years_count = df.groupby(['platform','year_of_release']).count()
life = pd.DataFrame(list(years_count.index),columns=['platform','year']).groupby('platform')['year'].count()
my_hist(life,{'title':'Время жизни платформы','bins':len(life.unique())})
pl_life = life.describe()['mean']
print('Игровые платформы существуют в среднем {:.2f} лет.'.format(pl_life),\
      'Для PC игры выходят с 1985 года, но это исключение.')
```

Отбросим данные по платформам,для которых игры не выходят уже 7 лет (те с 2010г)

```python
actual_plm = df.query('year_of_release > {:}'.format(2016 - math.floor(pl_life)))['platform']
df_new = df.loc[df['platform'].isin( actual_plm.unique() )]
display(pivot_table_pl_sum_by_years(df_new))
```

Отбросим данные по платформам,для которых игры уже не вышли в 2016 году
DS, PS2, PSP. Они прожили уже свои 7+ лет, по таблице видно, что продажи упали еще до 16 года. 
Вышло следующее поколение платформ, например SP3, SP4

```python
actual_plm = df.query('year_of_release == 2016')['platform']
df_new = df_new.loc[df['platform'].isin( actual_plm.unique() )]
display(pivot_table_pl_sum_by_years(df_new))
```

Из <a href="#gr">графиков</a> выше и таблицы видим, что настоящее развитие игр началось в 2000-х, все современные платформы (кроме PC, конечно) посвились после 2004. Удалим данные до 2005 года.

```python
df_new = df_new.query('year_of_release >= 2005')
display(pivot_table_pl_sum_by_years(df_new))
```

Платформы с наибольшими суммарными продажами и распределение по годам.

```python
def top_values(data,col,plot_title):    
    m = data.pivot_table(index=col, values=['sum_sales'], aggfunc='sum')
    m.columns = ['Количество копий (млн)']
    m.index.name = None
    display(m.sort_values(by='Количество копий (млн)', ascending=False))
    i = 'Количество копий (млн)'
    my_bar(m[i],plot_title,'Количество копий (млн)')
    m_view = pd.concat([m[i], round((m[i]/sum(m[i]))*100,2)], axis=1,sort=False)
    m_view.columns = ['К-во проданных копий','Доля рынка (%)' ]
    top_platforms = m_view.sort_values(by='К-во проданных копий', ascending=False).head(5)
    print('Top-5')
    display(top_platforms)
    return list(top_platforms.index)

print("Игры 2005 - 2016".upper())
top_pl_keys = top_values(df_new,'platform','Сумма проданных игр по платформам (2005 - 2016)')
print('')
print("Новые игры (2016 года)".upper())
top_pl_keys_16 = top_values(df_new.query('year_of_release == 2016'),'platform',\
                            'Сумма проданных игр по платформам (2016)')
```

```python
def plot_by_tariff(platform):
    a = df_new.query('platform =="' + platform + '"').groupby('year_of_release')['name'].count()
    my_bar(a, 'Количество игр для платформы ' + platform + ' по годам', 'Количество игр')

for platform in top_pl_keys:
    plot_by_tariff(platform)
```

* Из графиков видно, что только PS4 набирает популярность. Число игр под эту платформу растет последние 4 года и, судя по средней продолжительности жизни платформы, PS4 будет еще пользоваться популярностью в ближайшие годы
* Сумма проданых копий падает, но это не всегда означает, что платформа отмирает, так как из <a href="#gr">графиков</a> видно, что падает количество новых игр после 2009 года в целом.
* Жизнь платформ имеет в целом нормальное распределение. По таблицам и графикам видно, что, хотя Wii, X360 и PS3 и попали в топ платформ, для которых выпущено максимальное количество игр в разные годы, но они давно прошли свой максимум, и их дни сочтены.
* XOne появилась не так давно и не попала по количеству игр в топ, зато в 2016 заняла второе место после PS4

Возьмем данные за время жизни самой популярной платформа SP4 ( те с 2013 года)

```python
game_df = df_new.query('year_of_release >= 2013')        
print("Платформы, для которых продано больше всего копий игр, выпущенных после выхода PS4".upper())
top_platforms = top_values(game_df,'platform','Количество выпущенных игр по платформам') 
```

Таким образом, можно выделить первую тройку самых популярных платформ: PS4, XOne и 3DS

<b>«Ящик с усами» по глобальным продажам игр в разбивке по платформам.</b> <a id='data_analiz3'></a>

Отрежем выбросы, чтобы график выглядел "читабельно"

```python
quant = []
for i in top_platforms:
    quant.append(game_df.loc[df_new['platform'] == i]['sum_sales'].quantile([0.95])[0.95])

data = game_df.query('sum_sales < {:.2f}'.format(min(quant))).loc[df['platform'].isin(top_platforms)]
a = sns.boxplot(x="platform", y="sum_sales",  data=data)
```

* Количество проданных экземпляров игры для всех популярных платформ о самом общем случае не превышает 1.25 млн, медианное среднее в районе 0.2 млн. Однако, всех популярных платформ есть множество очень популярных игр проданных в количестве более 1,5 млн

<b>Корреляция отзывов и продаж для SP4</b><a id="data_analize4"></a>

```python
data = game_df.query('platform == "PS4"')
display(data.corr().style.background_gradient('coolwarm'))

plt.scatter(x = data['critic_score'], y = data['sum_sales'], c = data['sum_sales']) 
plt.ylabel('Количество проданных игр(млн)') 
plt.xlabel('Оценка критиков (0 - 100)') 
plt.title('Диаграмма рассеяния оценки критиков и количества проданных игр') 
plt.grid()
plt.show()

plt.scatter(x = data['user_score'], y = data['sum_sales'], c = data['sum_sales']) 
plt.ylabel('Количество проданных игр(млн)') 
plt.xlabel('Оценка пользователей (0 - 10)') 
plt.title('Диаграмма рассеяния оценки пользователей и количества проданных игр') 
plt.grid()
plt.show()
```

* Продаваемость игры не так сильно зависит от оценки критиков (корреляция 0,2). Есть игры с высокими оценками и маленькими продажами
* Так как коэффициент корреляции зависимости суммы продаж от оценки пользователей по модулю всего 0.1, то можно сказать, что успех игры не связан с оценкой пользователей. По диаграмме рассеяния видно, что большинство игр имеют неплохие отзывы и при этом низкие продажи, но есть также игры с баллом меньше 6 и высокими продажами. ### Статистический анализ <a id="stat"></a>

<b>Средние пользовательские рейтинги платформ Xbox One и PC одинаковые</b><a id='stat1'></a>

<ul>
        <li><b>Нулевая гипотеза Н0</b> = "Средние пользовательские рейтинги платформ Xbox One и PC одинаковые"</li>
        <li><b>Альтернативная гипотеза H1</b> = "Средние пользовательские рейтинги платформ Xbox One и PC различаются"</li>
    </ul>
    
- Критический уровень статистической значимости 0.05
- Используем ttest_ind для 2 независимых выброк
Так как дисперсии отличаются больше, чем на 5%, выбираем equal_var = False

```python
from scipy import stats as st

def diff_var(v1,v2):
    a = v1.var(ddof=1)      
    b = v2.var(ddof=1)
    if a > b:
        print('Дисперсии выборок отличаются на {:.2%}'.format((a - b)/a))
    else:
        print('Дисперсии выборок отличаются на {:.2%}'.format((b - a)/b))
    
xb0 = game_df.query('platform == "XOne"')
pc = game_df.query('platform == "PC"')
alpha = 0.05

my_hist(xb0['user_score'],{'title':'Пользовательский рейтинг Xbox One','bins':10})
display(xb0['user_score'].describe())

my_hist(pc['user_score'],{'title':'Пользовательский рейтинг PC','bins':10})
display(pc['user_score'].describe())

diff_var(xb0['user_score'],pc['user_score'])
results= st.ttest_ind(xb0['user_score'], pc['user_score'], equal_var = False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
    print('Средние пользовательские рейтинги платформ Xbox One и PC различаются')
else:
    print('Не получилось отвергнуть гипотезу')
    print("Средние пользовательские рейтинги платформ Xbox One и PC одинаковые")
```

<b>Средние пользовательские рейтинги жанров Action и Sports разные</b><a id='stat2'></a>

<ul>
    <li><b>Нулевая гипотеза Н0</b> = "Средние пользовательские рейтинги жанров Action и Sports одинаковые"</li>
    <li><b>Альтернативная гипотеза H1</b> = "Средние пользовательские рейтинги жанров Action и Sports различаются"</li>
</ul>

- Критический уровень статистической значимости 0.05
- Используем ttest_ind для 2 независимых выброк
Так как дисперсии отличаются больше, чем на 5%, выбираем equal_var = False

```python
action = game_df.query('genre == "Action"')
sports = game_df.query('genre == "Sports"')
alpha = 0.05 

my_hist(action['user_score'],{'title':'Пользовательский рейтинг жанра "Action"','bins':10})
display(action['user_score'].describe())

my_hist(sports['user_score'],{'title':'Пользовательский рейтинг жанра "Sports"','bins':10})
display(sports['user_score'].describe())

diff_var(action['user_score'],sports['user_score'])

results= st.ttest_ind(action['user_score'],sports['user_score'], equal_var = False)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
    print('Средние пользовательские рейтинги жанров Action и Sports различаются')
else:
    print('Не получилось отвергнуть гипотезу')
    print("Средние пользовательские рейтинги жанров Action и Sports одинаковые")
```

<b>Вывод</b><a id='stat4'></a>

* Платформа Xbox One входит в топ-5 в Европе и Северной Америке. Для нее, также, так для PC выходят игры всех жанров, кроме Puzzle, однако PC всегда найдет своих пользователей. Средний бал пользовательских оценок игр у этих двух платформ примерно одинаковый и может лишь случайно отличатся с вероятностью 5%
* Рейтинги жанров Action и Sports различаются. Медианное среднее отличается больше,чем на бал, значение pvalue очень мало, поэтому разница значительна. Игры жанра Action самые популярные в Европе и Северной Америке, количество игр этого жанра превышает количество игр во всех других жанрах и они выходят под все платформы - эти игры самые дорогие и качественные, поэтому набирают много хороших отзывов
